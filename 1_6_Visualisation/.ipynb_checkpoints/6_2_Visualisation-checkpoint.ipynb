```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

## Логарифмические масштабы

```python
x = np.linspace(-5, 5, 100)

plt.figure()
plt.plot(x, np.exp(x) + np.exp(-x))
plt.yscale('log')
plt.yticks(fontsize=15)
plt.show()
```

```python
x = np.logspace(-2, 2, 100)

plt.figure()
plt.plot(x, x + x**3)
plt.xscale('log'), plt.xticks(fontsize=12)
plt.yscale('log'), plt.yticks(fontsize=12)
plt.show()
```

### Спираль Архимеда

```python
t = np.linspace(0, 2*np.pi, 100)

plt.figure()
plt.polar(t, t)
plt.show()
```

### Кардиода

```python
phi = np.linspace(0, 2*np.pi, 400)

plt.figure()
plt.polar(phi, 2 * (1 - np.cos(phi)))
plt.show()
```

## Контурные графики

```python
x = np.linspace(-1, 1, 50)
y = x
z = np.outer(x, y)

plt.figure(figsize = (5, 5))
plt.contour(x, y, z)
plt.show()
```

```python
plt.figure(figsize = (5, 5))
curves = plt.contour(x, y, z, np.linspace(-1, 1, 11))
plt.clabel(curves)
plt.title(r'$z=xy$', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (5, 5))
plt.contourf(x, y, z, np.linspace(-1, 1, 11))
plt.colorbar()
plt.show()
```

## Трехмерные линии

```python
t = np.linspace(0, 4*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t / (4*np.pi)
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure = False)
fig.add_axes(ax)
ax.plot(x, y, z)
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure = False)
ax.elev, ax.azim = 45, 30
fig.add_axes(ax)
ax.plot(x, y, z)
plt.show()
```

## Поверхности

```python
X = 10
N = 50
u = np.linspace(-X, X, N)
x, y = np.meshgrid(u, u)
r = np.sqrt(x**2 + y**2)
z = np.sin(r) / r

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure = False)
ax.plot_surface(x, y, z, rstride = 10, cstride=10)
fig.add_axes(ax)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure = False)
ax.plot_surface(x, y, z, rstride = 1, cstride=1, cmap = 'gnuplot')
fig.add_axes(ax)
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 50)
theta, phi = np.meshgrid(t, t)
r = 0.2
x, y, z = (
            (1 + r * np.cos(phi)) * np.cos(theta),
            (1 + r * np.cos(phi)) * np.sin(theta),
            r * np.sin(phi)
            )

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure = False)
ax.elev = 60
ax.plot_surface(x, y, z, rstride = 2, cstride=1, cmap = 'gnuplot')
fig.add_axes(ax)
plt.show()
```

## Recursion (отвлекся типа) - расчет сложных процентов по вкладам

```python
def proc(x, n):
    if n == 1: 
        return (1 + x/100)
    else:
        return proc(x, n-1) * (1 + x / 100)

y = proc(3,100)
print(f'{y:5f}')
```
