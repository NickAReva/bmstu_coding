```python
for dtype in ['float', 'int', 'object']:
    df_selected_type = df.select_dtypes(include=dtype)
    mean_usage = df_selected_type.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage/1024 ** 2
    print(f"Average memory usage for {dtype} columns: {mean_usage_mb:.3f}")
```
