Создайте класс **окружность** с атрибутом **радиус**. В классе должны быть следующие методы:
* расчет диаметра
* расчет длины окружности
* расчет площади

Продемонстрируйте правильность работы кода

```python
class circle:
    def __init__(self, radius):
        self.radius = radius

    def diameter(self):
        return self.radius * 2

    def length(self):
        return 2 * np.pi * self.radius

    def area(self):
        return np.pi * self.radius **2
```

### Задача 2

Создайте класс **прямоугольник** с атрибутами **длина** и **ширина**. В классе должны быть следующие методы:
* расчет площади прямоугольника
* расчет диагонали прямоугольника
* расчет радиуса описанной вокруг прямоугольника окружности
* расчет площади описанной окружности
* расчет площади вписанной окружности
* расчет длины окружности вписанной окружности
* расчет длины окружности описанной окружности



```python
class rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    #расчет площади прямоугольника
    def area(self):
        return self.length * self.width

    #расчет диагонали прямоугольника
    def diagonal(self):
        return (self.length**2 + self.width**2)**0.5

    #расчет радиуса описанной вокруг прямоугольника окружности
    def radius_circumscribed(self):
        return self.diagonal() / 2

    #расчет радиуса вписанной в прямоугольник окружности
    def radius_inscribed(self):
        if self.length > self.width:
            return self.width / 2
        else:
            return self.length / 2

    #расчет площади описанной окружности
    def area_circumscribed(self):
        return np.pi * self.radius_circumscribed()**2

    #расчет площади вписанной окружности
    def area_inscribed(self):
        return np.pi * self.radius_inscribed()**2

    #расчет длины окружности вписанной окружности
    def legth_circumscribed(self):
        return 2 * np.pi * self.radius_circumscribed()
    #расчет длины окружности описанной окружности
    def legth_inscribed(self):
        return 2 * np.pi * self.radius_inscribed()
```

```python
rec = rectangle(1.414213562373095, 1.414213562373095)
#расчет площади прямоугольника
print(rec.area())
#расчет диагонали прямоугольника
print(rec.diagonal())
#расчет радиуса описанной вокруг прямоугольника окружности
print(rec.radius_circumscribed())
#расчет радиуса вписанной в прямоугольник окружности
print(rec.radius_inscribed())
#расчет площади описанной окружности
print(rec.area_circumscribed())
    #расчет площади вписанной окружности
print(rec.area_inscribed())
    #расчет длины окружности вписанной окружности
print(rec.legth_circumscribed())
    #расчет длины окружности описанной окружности
print(rec.legth_inscribed())
```

### Задача 2: Вычисления по цепочке

Напишите функцию, которая вычисляет значение следующей математической функции в зависимости от переменной х, где х - вектор

$F(x) = \frac{3}{4}AA^Tx + Bx + c$


```python
np.random.seed(42)

A = np.random.randint(low=1, high=5, size=(3,3))
B = np.random.randint(low=1, high=5, size=(3,3))
c = np.random.randint(low=1, high=5, size=(3))

def F(x):
    A1 = (3 / 4) * np.dot(A, A.transpose())
    AX = A1 @ x
    BX = B @ x
    return AX + BX + c

print(F(x))
```

### Задача 3:

Дана разреженная матрица, преобразуйте ее в csr формат, посчитайте долю разреженности, т.е. содержания нулей

```python
from scipy.sparse import csr_matrix


# фиксируем сид
np.random.seed(42)

A = np.eye(50)
A[5, :] = np.random.randint(low=0, high=2, size = (50))
A[31, :] = np.random.randint(low=0, high=2, size = (50))
A[:, 49] = np.random.randint(low=0, high=2, size = (50))

B = csr_matrix(A)
(a, b) = B.shape
c = np.count_nonzero(B.toarray())
(a * b - c) / (a * b)
```

### Задача 4

Даны две разреженные матрицы, преобразуйте их в формат csr и посчитайте:

* сумму двух матриц
* матричное произведение двух матриц
* выведите на экран сумму и произведение, предварительно преобразовав их обратно в "плотный" формат (метод .todense)

```python
# фиксируем сид
np.random.seed(42)

A = np.eye(50)
A[2, :] = np.random.randint(low=0, high=2, size = (50))
A[37, :] = np.random.randint(low=0, high=2, size = (50))
A[:, 44] = np.random.randint(low=0, high=2, size = (50))

B = np.zeros((50,50))
B[2, :] = np.random.randint(low=0, high=2, size = (50))
B[37, :] = np.random.randint(low=0, high=2, size = (50))
B[:, 44] = np.random.randint(low=0, high=2, size = (50))
```

```python
A_csr = csr_matrix(A)
B_csr = csr_matrix(B)
AplusB_csr = A_csr + B_csr
AmultiB_csr = A_csr @ B_csr
print('Sum of A and B matrices\n', AplusB_csr.todense())
print('Multiplication of A and B matrices\n', AmultiB_csr.todense())
```
