```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

## Простые графики

```python
plt.figure()
plt.plot([0, 1, 0.5])
plt.show()
```

```python
plt.figure()
plt.plot([0, 0.25, 1], [0, 1, 0.5])
plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 100)

plt.figure()
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.figure()
plt.scatter([0, 0.25, 1], [0, 1, 0.5])
plt.show()
```

## Параметрические функции

```python
t = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize = (4,4))
plt.plot(np.cos(t), np.sin(t))
plt.show()
```

## plot with multiple functions

```python
x = np.linspace(0, 4, 10)

plt.figure(figsize = (8,4))
plt.plot(x, x, 'r-', x, x**2, 'b--', x, x**0.5, 'k-.')
plt.show()
```

```python
t = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.figure(figsize = (8, 4))
plt.plot(t, np.sin(t), 'r--')
plt.plot(t, np.cos(t), 'c-.')
plt.show()
```

```python
x = np.linspace(0, 1, 11)

plt.figure()
plt.plot(x, x**2, 'ro')
plt.plot(x, 1 - x, 'gs')
plt.show()
```

## Настройка графиков

```python
x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize = (10,5))
plt.plot(x, np.sin(x), 
         linewidth =2, color = 'g', dashes = [8, 4],
         label = r'$\sin x$')
plt.plot(x, np.cos(x), 
         linewidth =2, color = 'r', dashes = [8, 4, 2, 4],
         label = r'$\cos x$')
plt.axis([0, 2*np.pi, -1, 1])
plt.xticks(np.linspace(0, 2*np.pi, 9), ('0', r'$\frac{1}{4}\pi$', r'$\frac{1}{2}\pi$',
                                       r'$\frac{3}{4}\pi$', r'$\pi$', 
                                       r'$1\frac{1}{4}\pi$', r'$1\frac{1}{2}\pi$', 
                                       r'$1\frac{3}{4}\pi$', r'$2\pi$'),
                                          fontsize = 20)
plt.yticks(fontsize = 12)
plt.xlabel(r'$x$', fontsize = 20)
plt.ylabel(r'$y$', fontsize = 20)
plt.title(r'$\sin x$, $\cos x$', fontsize =20)
plt.legend(fontsize = 20, loc = 0)

plt.show()
```

```python
x = np.linspace(-2, 2, 100)

plt.figure(figsize = (5, 5))
plt.plot(x, x**3, linestyle = '--', lw =2, label = '$y=x^3$')
plt.xlabel('x'), plt.ylabel('y')
plt.legend()
plt.title('График кубической функции')
plt.grid(ls=':')

plt.show()
```
