```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

## Логарифмические масштабы

```python
x = np.linspace(-5, 5, 100)

plt.figure()
plt.plot(x, np.exp(x) + np.exp(-x))
plt.yscale('log')
plt.yticks(fontsize=15)
plt.show()
```

```python
x = np.logspace(-2, 2, 100)

plt.figure()
plt.plot(x, x + x**3)
plt.xscale('log'), plt.xticks(fontsize=12)
plt.yscale('log'), plt.yticks(fontsize=12)
plt.show()
```

### Спираль Архимеда

```python
t = np.linspace(0, 2*np.pi, 100)

plt.figure()
plt.polar(t, t)
plt.show()
```

### Кардиода

## Recursion (отвлекся типа) - расчет сложных процентов по вкладам

```python
def proc(x, n):
    if n == 1: 
        return (1 + x/100)
    else:
        return proc(x, n-1) * (1 + x / 100)

y = proc(3,100)
print(f'{y:5f}')
```
